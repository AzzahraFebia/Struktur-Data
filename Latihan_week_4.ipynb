Implementasi queue dengan LIST

```python
# Implementasi Queue menggunakan List
class QueueList:
  def __init__(self):
    self.queue = []

  def enqueue(self, item):
    self.queue.append(item)

  def dequeue(self):
    if not self.is_empty():
      return self.queue.pop(0)
    return None

  def front(self):
    return self.queue[0] if not self.is_empty() else None

  def rear(self):
    return self.queue[-1] if not self.is_empty() else None

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)

# contoh
queue = QueueList()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print("Queue setelah enqueue:", queue.queue)
print("Dequeue element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue Size:", queue.size())
```

Implementasi Queue dengan collections.deque

```python
from collections import deque

# Implementasi Queue menggunakan deque
class QueueDeque:
  def __init__(self):
    self.queue = deque()

  def enqueue(self, item):
    self.queue.append(item)

  def dequeue(self):
    return self.queue.popleft() if not self.is_empty() else None

  def front(self):
    return self.queue[0] if not self.is_empty() else None

  def rear(self):
    return self.queue[-1] if not self.is_empty() else None

  def is_empty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)

 # contoh
queue = QueueDeque()
queue.enqueue(7)
queue.enqueue(15)
queue.enqueue(20)
print("Queue setelah enqueue:", list(queue.queue))
print("Dequeue element:", queue.dequeue())
print("Front element:", queue.front())
print("Rear element:", queue.rear())
print("Queue Size:", queue.size())
```

Implementasi Queue dengan Linked List

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class QueueLinkedList:
  def __init__(self):
    self.front = self.rear = None

  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
      self.front = self.rear = new_node
      return
    self.rear.next = new_node
    self.rear = new_node

  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    return dequeued_element

  def is_empty(self):
    return self.front is None

  def front_element(self):
    return self.front.data if self.front else None

  def rear_element(self):
    return self.rear.data if self.rear else None

# contoh
queue = QueueLinkedList()
queue.enqueue(15)
queue.enqueue(27)
queue.enqueue(6)
print("Dequeue element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

Antrian Pemesanan Tiket Bioskop

```python
class AntrianTiket:
  def __init__(self, kapasitas=12):
    self.antri = []
    self.kapasitas = kapasitas

  def enqueue(self, ticket_id): # Menambah pelanggan
    if len(self.antri) < self.kapasitas:
      if '7' in ticket_id:
        self.antri.insert(0, ticket_id)
        print(f"Customer {ticket_id} is VIP, silahkan menempati kursi anda!")
      else:
        self.antri.append(ticket_id)
        print(f"Customer {ticket_id} masuk antrian.")
        self.kapasitas -= 1 # Mengurangi kapasitas untuk pelanggan non-vip
    else:
      print("Full, please wait.")

  def dequeue(self): # Melayani pelanggan (elemen keluar)
    if not self.is_empty():
      Customer = self.antri.pop(0)
      print(f"Customer {Customer} keluar antrian.")
      self.kapasitas += 1 # Menambah kapasitas jika pelanggan keluar
    else:
      print("Empty.")
      self.kapasitas = 12 # Kapasitas kembali 12 ketika antrian kosong

  def front(self): # Melihat pelanggan yang masuk pertama
    if not self.is_empty():
      return self.antri[0]
    else:
      return None

  def rear(self): # Melihat pelanggan yang masuk terakhir
    if not self.is_empty():
      return self.antri[-1]
    else:
      return None

  def is_empty(self): # Mengecek apakah antrian kosong
    return len(self.antri) == 0

  def size(self): # Menghitung jumlah pelanggan dalam antrian
    return len(self.antri)

# PENGGUNAAN
antri = AntrianTiket()

print("\n---- ANTRIAN TIKET BIOSKOP ----")
while True:
  print("\nMENU:")
  print("1. Add Customer")
  print("2. Serve Customer")
  print("3. Lihat kapasitas antrian")
  print("4. Lihat pelanggan pertama")
  print("5. Lihat pelanggan terakhir")
  print("6. Out")
  choice = input("Enter your choice: ")

  if choice == '1':
    ticket_id = input("Enter customer ticket id: ")
    antri.enqueue(ticket_id)
  elif choice == '2':
    antri.dequeue()
  elif choice == '3':
    print("Capacity:", antri.kapasitas)
  elif choice == '4':
    print("Front:", antri.front())
  elif choice == '5':
    print("Rear:", antri.rear())
  elif choice == '6':
    print("Thank you.")
    break
  else:
    print("Invalid. Please try again.")
```

Antrian Layanan Fotokopi Kampus

```python
class AntriFotokopi:
  def __init__(self, kapasitas=10):
    self.antri = []
    self.kapasitas = kapasitas

  def enqueue(self, order_id, num_pages): # Menambah pesanan
    if len(self.antri) < self.kapasitas:
      if num_pages < 10:  # Fast Track
          print(f"Pesanan Fast Track {order_id} akan diproses!")
      else:
          self.antri.append((order_id, num_pages))
          print(f"Pesanan {order_id} ditambah ke antrian.")
          self.kapasitas -= 1 # Pengurangan kapasitas ketika ada pesanan baru
    else:
        print("Full, please wait.")

  def dequeue(self): # Menyelesaikan pesanan
    if not self.is_empty():
      order_id, num_pages = self.antri.pop(0)
      print(f"Pesanan {order_id} sebanyak {num_pages} halaman akan diproses!")
      self.kapasitas += 1 # Penambahan kapasitas ketika pesanan selesai
      if num_pages > 100:
       self.enqueue(order_id, num_pages - 100) # Memproses 100 halaman di 2 tahap
      print(f"Pesanan {order_id} telah diproses sebagian, sisa halaman berada di antrian.")
    else:
      print("Empty.")

  def front(self): # Melihat pesanan pertama tanpa mengeluarkannya
    if not self.is_empty():
      return self.antri[0]
    else:
      return None

  def rear(self): # Melihat pesanan terakhir tanpa mengeluarkannya
    if not self.is_empty():
      return self.antri[-1]
    else:
      return None

  def is_empty(self): # Memeriksa apakah antrian kosong
     return len(self.antri) == 0

  def size(self): # Jumlah pelanggan dalam antrian
     return len(self.antri)

# PENGGUNAAN
antri = AntriFotokopi()

print("\n>>>> LAYANAN FOTOKOPI <<<<")
while True:
    print("\nMenu:")
    print("1. Add Order")
    print("2. Serve")
    print("3. Lihat Pesanan Pertama")
    print("4. Lihat Pesanan Terakhir")
    print("5. Lihat Kapasitas Antrian")
    print("6. Out")

    choice = input("Enter your choice: ")

    if choice == '1':
        order_id = input("Enter ID: ")
        num_pages = int(input("Enter pages: "))
        antri.enqueue(order_id, num_pages)
    elif choice == '2':
        antri.dequeue()
    elif choice == '3':
      print("First order:", antri.front())
    elif choice == '4':
      print("Last order:", antri.rear())
    elif choice == '5':
        print("Capacity:", antri.kapasitas)
    elif choice == '6':
        print("Thank you.")
        break
    else:
        print("Invalid.")
```

Antrian Layanan Parkir Mall

```python
class AntrianParkir:
  def __init__(self, kapasitas=15):
    self.antri = []
    self.kapasitas = kapasitas

  def enqueue(self, vehicle_number, priority): # Menambah Kendaraan di Parkiran
    if len(self.antri) < self.kapasitas:
      if priority.lower() == 'ya': # Parkir Prioritas
        self.antri.insert(0, vehicle_number)
        print(f"Kendaraan listrik {vehicle_number} is priority, silahkan masuk.")
        self.kapasitas -= 1 # Kapasitas Parkiran Berkurang Ketika Kendaraan Masuk
      else:
        self.antri.append(vehicle_number)
        print(f"Kendaraan {vehicle_number} masuk parkiran.")
        self.kapasitas -= 1 # Kapasitas Parkiran Berkurang Ketika Kendaraan Masuk
    else:
      print("Parkiran penuh, please wait!")

  def dequeue(self): # Kendaraan Keluar
    if not self.is_empty():
      vehicle_number = self.antri.pop(0)
      print(f"Kendaraan {vehicle_number} keluar.")
      self.kapasitas +=1 # Kapasitas Parkiran Bertambah Ketika Kendaraan Masuk
    else:
      print("Empty.")

  def front(self): # Melihat Kendaraan Pertama Yang Masuk
    if not self.is_empty():
      return self.antri[0]
    else:
      return None

  def rear(self): # Melihat Kendaraan Terakhir Yang Masuk
    if not self.is_empty():
      return self.antri[-1]
    else:
      return None

  def is_empty(self): # Memeriksa Apakah Parkiran Kosong
      return len(self.antri) == 0

  def size(self): # Jumlah Kendaraan
      return len(self.antri)

# PENGGUNAAN
antri = AntrianParkir()

print("\n.... LAYANAN PARKIR ....")
while True:
    print("\nMenu:")
    print("1. Kendaraan Masuk")
    print("2. Kendaraan Keluar")
    print("3. Lihat Kendaraan Pertama")
    print("4. Lihat Kendaraan Terakhir")
    print("5. Lihat Kapasitas")
    print("6. Lihat Jumlah Kendaraan")
    print("7. Out")

    choice = input("Enter your choice: ")

    if choice == '1':
        vehicle_number = input("Input nomor plat: ")
        priority = input("Apakah kendaraan listrik? (ya/tidak): ")
        antri.enqueue(vehicle_number, priority)
    elif choice == '2':
        antri.dequeue()
    elif choice == '3':
        print("Kendaraan pertama:", antri.front())
    elif choice == '4':
        print("Kendaraan terakhir:",antri.rear())
    elif choice == '5':
        print("Capacity:", antri.kapasitas)
    elif choice == '6':
        print("Jumlah kendaraan:", antri.size())
    elif choice == '7':
        print("Thank you.")
        break
    else:
        print("Invalid.")
```

Antrian Penambilan Sertifikat di Kampus

```python
class AntrianSertifikat:
  def __init__(self, kapasitas=20):
    self.antri = []
    self.kapasitas = kapasitas

  def enqueue(self, student_id, is_prestasi=False): # Menambah Antrian Mengambil Sertifikat
    if len(self.antri) < self.kapasitas:
      if is_prestasi.lower() == 'ya':
        self.antri.insert(0, student_id)
        print(f"Mahasiswa berprestasi {student_id} silahkan mengambil sertifikat.")
        self.kapasitas -= 1 # Kapasitas Antrian Berkurang Ketika Mahasiswa Bertambah
      else:
        self.antri.append(student_id)
        print(f"Mahasiswa {student_id} masuk antrian.")
    else:
      print("Antrian penuh, please wait.")

  def dequeue(self): # Menyelesaikan Pengambilan Sertifikat
     if not self.is_empty():
      student_id = self.antri.pop(0)
      print(f"Mahasiswa {student_id} mengambil sertifikat.")
      self.kapasitas += 1 # Kapasitas Antrian Bertambah Ketika Mahasiswa Berkurang
     else:
      print("Empty.")

  def front(self): # Menampilkan antrian pertama tanpa mengeluarkannya
     if not self.is_empty():
      return self.antri[0]
     else:
      return None

  def rear(self): # Menampilkan antrian terakihir tanpa mengeluarkannya
     if not self.is_empty():
      return self.antri[-1]
     else:
      return None

  def is_empty(self): # Memeriksa apakah antrian kosong
      return len(self.antri) == 0

  def size(self): # Jumlah antrian
      return len(self.antri)

# PENGGUNAAN
antri = AntrianSertifikat()

print("\n---- PENGAMBILAN SERTIFIKAT ----")
while True:
    print("\nMenu:")
    print("1. Masuk Antrian")
    print("2. Ambil Sertifikat")
    print("3. Lihat Mahasiswa Pertama")
    print("4. Lihat Mahasiswa Terakhir")
    print("5. Lihat Jumlah Mahasiswa")
    print("6. Out")

    choice = input("Masukkan pilihan Anda: ")

    if choice == '1':
        student_id = input("Enter nomor antrian: ")
        is_prestasi = input("Berprestasi? (ya/tidak): ")
        antri.enqueue(student_id, is_prestasi)
    elif choice == '2':
        antri.dequeue()
    elif choice == '3':
        print("Mahasiswa pertama:",antri.front())
    elif choice == '4':
        print("Mahasiswa terakhir:", antri.rear())
    elif choice == '5':
        print("Jumlah mahasiswa:", antri.size())
    elif choice == '6':
        print("Thank you.")
        break
    else:
        print("Invalid.")
```
