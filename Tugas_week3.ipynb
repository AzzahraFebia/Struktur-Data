```python
class gudang:
  def __init__(self, max):
    self.max = max
    self.tumpukan = []

  # Menambahkan barang ke gudang
  def push(self, barang):
    if len(self.tumpukan) <= self.max:
      self.tumpukan.append(barang)
      print(f"{barang} berhasil ditambah ke gudang!")
    else:
      print(f"Gudang full! barang tidak dapat ditambah.") # Pesan kesalahan ketika kapasitas gudang max

  # Menghapus atau mengambil barang dari gudang
  def pop(self):
    if self.tumpukan:
      barang = self.tumpukan.pop()
      print(f"{barang} berhasil diambil dari gudang.")
      return barang
    else:
      print("Tidak ada barang! Gudang kosong.") # Pesan kesalahan ketika gudang kosong
      return None

  # Menampilkan isi gudang
  def tampilkan_gudang(self):
    if self.tumpukan:
      print("Barang di gudang:")
      for barang in self.tumpukan:
        print(f"- {barang}")
    else:
      print("Tidak ada barang! Gudang kosong.")


# Membuat objek gudang
gudang = gudang(max)

print("---Simulasi Tumpukan Barang dalam Gudang---")

# Meminta input kapasitas gudang
kapasitas_max = int(input("\nMasukkan kapasitas maksimum gudang: "))

while True:
  print("\nMenu:")
  print("1. Tambah Barang")
  print("2. Ambil Barang")
  print("3. Tampilkan Isi Gudang")
  print("4. Out")

  opsi = input("Masukkan pilihan: ")

  if opsi == "1":
    barang = input("Masukkan nama barang: ")
    gudang.push(barang)
  elif opsi == "2":
    gudang.pop()
  elif opsi == "3":
    gudang.tampilkan_gudang()
  elif opsi == "4":
    print("Program selesai, see u next time!")
    break
  else:
    print("Invalid. Silakan coba lagi.")
```
